# Simulación

$$Serie + Tendencia + Estacionalidad + Ruido$$

```python
# Cargamos librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generamos los diferentes componentes de la serie
# Definimos la semilla
np.random.seed(73)

# Vamos a simular una serie mensual de 36 meses

# Primero establecemos los meses
meses = pd.date_range(start='2007-10-27',
                      periods = 36, freq = 'M')
# Tendencia
tendencia = np.linspace(100, 130, 36)

# Estacionalidad
estacionalidad = np.tile([10, 12, 15, 20, 25, 30, 28,
                          26, 22, 18, 15, 12], 3)

# Ruido
ruido =  np.random.normal(0, 2, 36)

# Serie
serie = tendencia + estacionalidad + ruido

# Armamos un DataFrame con toda la información
df = pd.DataFrame({'Fecha': meses,
                   'Serie': serie,
                   'Tendencia': tendencia,
                   'Estacionalidad': estacionalidad,
                   'Ruido': ruido})
df.head()
```

```python
#  Exportamos los datos a Excel

df.to_excel('datos.xlsx', index = False)
```
